## Conversion from SMILES to .pt
1. SMILESからptファイルを作成する
2. ファイル名が入ったtargets.csvを作成する（物性値は別途用意）

```python
import os
import pandas as pd
import random
from process import *
from summary import *
import torch
import shutil
from rdkit import Chem
```

```python
# Set X and y
index = "SMILES"
target = "Bandgap"#"Melting_point"

# Random sampling (Please set datasize)
path="D:datasets/OMDB-GAP1_v1.1.tar/OMDB-GAP1_v1.1/"
# df = pd.read_csv(path+"mp-data_no-disorder_clean.csv", index_col=0) 
df = pd.read_csv(path+"smiles-bandgap_clean.csv")

df_sample = df.sample(2500) 
df_sample = df_sample[[index, target]]

# Spedify folder
path_json="D:datasets/orgcrystmp_data_nodisorder/molgraph/json/Bandgap_addH/data2500_rand1/"
if not os.path.exists(path_json):
    os.mkdir(path_json)

# Make .csv
df_sample.to_csv(path_json+"targets.csv", header=None, index=None)

# Convert SMILES to .pt
if not os.path.exists(path_json+"raw/"):
    os.mkdir(path_json+"raw/")
if os.path.exists(path_json+"targets.csv"):
    shutil.move(path_json+"targets.csv", path_json+"raw/targets.csv")
Dataset = MoleculeDataset(root=path_json, filename="targets.csv", addH=True)

# Reset raw folder
shutil.move(path_json+"raw/targets.csv", path_json+"targets.csv")
os.rmdir(path_json+"raw/")
```

# Code for large size data of PubChem

```python
# Set X and y
index = "SMILES"
target = "Target"

# Batching
path = "D:datasets/orgcrystmp_data_nodisorder/testdata/"
df = pd.read_csv(path+"PubChem_clean.csv")
df = df[[index, target]]

cycle = int(df.shape[0]/10000)+1

for i in range(cycle):
    path_json="D:datasets/orgcrystmp_data_nodisorder/testdata/PubChem_molgraph_sub/"+str(i)+"/"
    print(path_json)
    if not os.path.exists(path_json):
        os.mkdir(path_json)
    
    start = i*10000
    end = (i+1)*10000
    
    if end > df.shape[0]:
        df_sample = df.iloc[start:, :]
    else:
        df_sample = df.iloc[start:end, :]
        
    # Make .csv
    df_sample.to_csv(path_json+"targets.csv", header=None, index=None)

    # Convert SMILES to .pt
    if not os.path.exists(path_json+"raw/"):
        os.mkdir(path_json+"raw/")
    if os.path.exists(path_json+"targets.csv"):
        shutil.move(path_json+"targets.csv", path_json+"raw/targets.csv")
    Dataset = MoleculeDataset(root=path_json, filename="targets.csv", addH=False)

    # Reset raw folder
    shutil.move(path_json+"raw/targets.csv", path_json+"targets.csv")
    os.rmdir(path_json+"raw/")
```
